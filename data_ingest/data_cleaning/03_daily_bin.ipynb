```python
import git
import os
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

repo = git.Repo('.', search_parent_directories=True)


os.chdir(repo.working_tree_dir)
```

```python
covari = pd.read_csv('data_ingest/data/modified/RF_ready_covari.csv')
print(covari.loc[40:60])
```

```python
print(len(covari))
```

```python
covari['time'] = pd.to_datetime(covari['time'])
covari.set_index('time', inplace=True)
covari_numeric = covari.drop('cruise', axis=1)
covari_cruise = covari['cruise'].copy()

daily_df = covari_numeric.resample('D').mean()
daily_df['cruise'] = covari_cruise.resample('D').first()
daily_df.reset_index(inplace=True)
```

```python
print(len(daily_df))
```

```python
daily_df.isna().sum() 
```

```python
daily_df.dropna(inplace=True)
print(len(daily_df))
```

```python
daily_df.to_csv('data_ingest/data/modified/covari_daily.csv', index=False)
```
